## Modelo de Regressão Linear

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

```python
Y = df['Precipitacao']
X = df[['TempMinima','Umidade Relativa Media','Velocidade do Vento Media','TempMaxima','Temp Comp Media']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=2811)
```

```python
 X_train.shape
```

```python
y_train.shape
```

```python
modelo = LinearRegression()
```

```python
modelo.fit(X_train, y_train)
```

```python
print("R quadrado = {}".format(modelo.score(X_train,y_train).round(2)))
```

```python
lm = modelo.predict(X_test)
```

```python
lm
```

```python
from sklearn.model_selection import cross_val_score
resultado = cross_val_score(modelo, X_test, y_test, cv = 10)
print(resultado.mean())
```
